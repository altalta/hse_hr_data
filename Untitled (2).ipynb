```python
import sqlite3
import pandas as pd
```

```python
# Создаем коннектор к файлу нашей базы данных. 
# Само собой, файлик Базы Данных должен лежать рядом с файлом блокнота (который вы сейчас читаете). 
con = sqlite3.connect('DBname.db')
```

```python
# Не спрашивайте
cur = con.cursor()
```

```python
cur.execute('SELECT DISTINCT Name FROM EmplScore').fetchall()
```

```python
cur.execute("""

    SELECT Name, COUNT (Score)
    FROM EmplScore
    WHERE Score = '5'
    GROUP BY Name
    ORDER BY Score

""").fetchall()
```

```python
df = pd.read_sql('SELECT DISTINCT Name FROM EmplScore', con)
```
